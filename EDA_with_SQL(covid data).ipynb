## Total cases and total deaths

```python
## total cases by location
%sql select location, sum(new_cases) as total_cases from covid group by location order by total_cases desc
```

In total, there are more than 200mln cases in the world, with 68mln in Asia, 55 in Europe and 46 in North America. The USA is the most infected country in the world with almost 39mln cases.

```python
#death rate per country
%sql select location, total_cases,total_deaths, round((total_deaths/total_cases)*100, 2) as death_rate from covid where date = (select max(date) from covid) order by 4 desc
```

It can be seen that the highest death rates are assigned to the countries with poor economic/social conditions(e.g. Sudan, Syria). The best result is demonstrated by Singapore with only 0.08% of deaths.

```python
## the same for continents
%sql select  continent, sum(new_cases),sum(new_deaths), round((sum(new_deaths)/sum(new_cases)),2) as death_rate from covid group by continent order by 4 desc
```

Obviously, Africa has the highest death percentage together with South America among the other continents

```python
#the global situation by each day
%sql select location, date, new_cases, new_deaths, round((new_deaths/new_cases),2) as death_rate from covid where location = 'World'

```

```python
#the worst days of COVID
%sql select location, date, new_cases, new_deaths, round((new_deaths/new_cases),2) as death_rate from covid where location = 'World' order by 5 desc limit 5

```

```python
#mean death rate per day
%sql select round(avg(round((new_deaths/new_cases),2)), 2) as mean_death_rate, avg(new_deaths) as mean_death_counts from covid where location = 'World'
```

The pandemic started at 22 of January 2020, whereas the worst week(by death cases) was in February 2020 when up to 24% of infected people died. Moreover, on average, 3% of new cases are dying everyday(which is approximately 7660 people). 

```python
#the day with the highest number of new infections
%sql select date, new_cases, new_deaths from covid where new_cases = (select max(new_cases) from covid as maximum_cases)
```

```python
#the day with the highest number of new deaths
%sql select date, new_deaths, new_cases from covid where new_deaths = (select max(new_deaths) from covid as maximum_deaths)
```

The highest increase in infected people was on 28 of April in 2021, while the highest increase in death counts was 3 months earlier, on 20 of January in 2021, almost 1 year later the pandemic began.

```python
%sql select location, sum(new_cases) as total_cases, sum(new_deaths) as total_deaths, round((sum(new_deaths)/sum(new_cases)), 2) as death_rate from covid where location like "%ussia%" group by location  
```

## Population and COVID-19

```python
#ratio of country's population to its covid cases and death cases
%sql select location, avg(population) as population, sum(new_cases) as cases, sum(new_deaths) as deaths, round(sum(new_deaths)/avg(population),5)*100 as death_percentage, round(sum(new_cases)/avg(population),5)*100 as cases_percentage from covid where continent is not null group by location order by 6 desc



```

In total, countries with small population are having pretty large percentage of people infected with COVID(like Seychelles, Andorra, San Marino and so on), however, they are not necessarily the worst on terms of death percentage, here Peru is the first one with apprx 0.6% of death cases from total population.

## Vaccination

```python
%sql select location, max(people_fully_vaccinated) as total_vac, round(( max(people_fully_vaccinated) / avg(population)),2) as vac_rate from covid where continent is not null group by location order by 3 desc 
```

Due to some potential data mistake, Gibraltar has more than 100% people vaccinated. The highest vaccinated rates are demonstrated by countries with small population. However,it is not always the case,for example,China has also pretty high rate - 62% of population.

```python
%sql select location, max(total_vaccinations) as total_vac from covid where continent is not null group by location order by 2 desc
```

Obviously, highly populated countries are the most vaccinated.
